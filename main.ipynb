```python


#Гистограмма для числовой переменной satisfaction_level:

sns.histplot(df['satisfaction_level'], kde=True)
plt.title('Распределение satisfaction_level')
plt.xlabel('satisfaction_level')
plt.ylabel('Частота')
plt.show()
     
```

```python


#Столбчатая диаграмма для категориальной переменной time_spend_company:
sns.countplot(x='time_spend_company', data=df)
plt.title('Распределение time_spend_company')
plt.xlabel('time_spend_company')
plt.ylabel('Количество')
plt.show()
     
```

```python


#Боксплот для числовой переменной last_evaluation:
sns.boxplot(x=df['last_evaluation'])
plt.title('Boxplot last_evaluation')
plt.xlabel('last_evaluation')
plt.show()
     
```



⏰

Распределение satisfaction_level:
    Распределен не нормальное. Вместо одного пика в центре, есть несколько пиков. Это означает, что есть несколько групп сотрудников с разной степенью удовлетворенности, а не одно равномерное распределение.

Распределение time_spend_company:
    Время, которое сотрудники провели в компании, также распределено не нормально. Большинство людей работают в компании 3 года. Нормальное распределение должно быть симметричным.

Boxplot last_evaluation:
    Оценки распределены довольно равномерно, но это не нормальное распределение. Нормальное распределение должно иметь один пик в центре и симметричный упадок графика в обе стороны.
